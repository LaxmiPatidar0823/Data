Pivot Table

```python
pd.pivot_table(data,'filled jobs','filled jobs diff',aggfunc=np.mean)
```

```python
pd.pivot_table(data,'filled jobs','filled jobs revised','filled jobs diff',aggfunc=np.mean)
```

```python
pd.pivot_table(data,'filled jobs','filled jobs revised','filled jobs diff',aggfunc=np.mean,margins=True)
```

```python
pd.pivot_table(data,'filled jobs',['filled jobs revised','filled jobs diff'],'total_earnings',aggfunc=np.mean,margins=True,margins_name="overall_Average")
```

```python
pd.pivot_table(data,'filled jobs','total_earnings',['filled jobs revised','filled jobs diff'],aggfunc=np.mean,margins=True,margins_name="overall_Average")
```
